```python
# Step 1: Input a DNA sequence
dna_seq = input("Enter a DNA sequence: ").upper()

# Step 2: Count each nucleotide
A_count = dna_seq.count('A')
T_count = dna_seq.count('T')
G_count = dna_seq.count('G')
C_count = dna_seq.count('C')

# Step 3: Calculate GC content
GC_content = ((G_count + C_count) / len(dna_seq)) * 100

# Step 4: Convert DNA to RNA
rna_seq = dna_seq.replace('T', 'U')

# Step 5: Print the results
print("\n--- DNA Analysis ---")
print(f"Length of sequence: {len(dna_seq)}")
print(f"A: {A_count}, T: {T_count}, G: {G_count}, C: {C_count}")
print(f"GC Content: {GC_content:.2f}%")
print(f"RNA sequence: {rna_seq}")
```

```python
# Step 6: Translate RNA to Protein
codon_table = {
    'AUG':'M', 'UUU':'F', 'UUC':'F', 'UUA':'L', 'UUG':'L',
    'CUU':'L', 'CUC':'L', 'CUA':'L', 'CUG':'L',
    'AUU':'I', 'AUC':'I', 'AUA':'I',
    'GUU':'V', 'GUC':'V', 'GUA':'V', 'GUG':'V',
    'UCU':'S', 'UCC':'S', 'UCA':'S', 'UCG':'S',
    'CCU':'P', 'CCC':'P', 'CCA':'P', 'CCG':'P',
    'ACU':'T', 'ACC':'T', 'ACA':'T', 'ACG':'T',
    'GCU':'A', 'GCC':'A', 'GCA':'A', 'GCG':'A',
    'UAU':'Y', 'UAC':'Y', 'UAA':'_', 'UAG':'_', 'UGA':'_',
    'CAU':'H', 'CAC':'H', 'CAA':'Q', 'CAG':'Q',
    'AAU':'N', 'AAC':'N', 'AAA':'K', 'AAG':'K',
    'GAU':'D', 'GAC':'D', 'GAA':'E', 'GAG':'E',
    'UGU':'C', 'UGC':'C', 'UGG':'W',
    'CGU':'R', 'CGC':'R', 'CGA':'R', 'CGG':'R',
    'AGU':'S', 'AGC':'S', 'AGA':'R', 'AGG':'R',
    'GGU':'G', 'GGC':'G', 'GGA':'G', 'GGG':'G'
}

protein = ''
for i in range(0, len(rna_seq)-2, 3):
    codon = rna_seq[i:i+3]
    protein += codon_table.get(codon, '-')

print(f"Protein sequence: {protein}")
```

```python
# Step 7: Visualize nucleotide composition
labels = ['A', 'T', 'G', 'C']
values = [A_count, T_count, G_count, C_count]

# Bar Chart
plt.figure(figsize=(8,4))
plt.bar(labels, values)
plt.title('Nucleotide Composition - Bar Chart')
plt.xlabel('Nucleotides')
plt.ylabel('Count')
plt.show()

# Pie Chart
plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Nucleotide Composition - Pie Chart')
plt.show()
```

```python
 #multiple DNA sequences (simulating samples)
dna_samples = {
    "Gene1": "ATGCGTACCTGA",
    "Gene2": "GCGTATCGTACG",
    "Gene3": "TTGCGGATCCGA",
    "Gene4": "ATATATCGCGCG"
}

#Analyze each DNA sequence
gc_values = []  # to store GC% of each sequence

for gene, seq in dna_samples.items():
    seq = seq.upper()
    A = seq.count('A')
    T = seq.count('T')
    G = seq.count('G')
    C = seq.count('C')
    GC = ((G + C) / len(seq)) * 100
    gc_values.append(GC)

    print(f"\n--- {gene} ---")
    print(f"Length: {len(seq)}")
    print(f"A:{A}, T:{T}, G:{G}, C:{C}")
    print(f"GC Content: {GC:.2f}%")
```

```python
# Step 3: Plot GC content comparison
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.bar(dna_samples.keys(), gc_values, color='teal')
plt.title('GC Content Comparison Between Genes')
plt.xlabel('Gene')
plt.ylabel('GC Content (%)')
plt.ylim(0, 100)
plt.show()
```
